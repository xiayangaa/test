```python
import math
x1,y1,x2,y2,x3,y3 = eval(input('请输入三组顶点，以（x,y）格式： '))
c = math.sqrt((x2-x1)**2+(y2-y1)**2)   #求边长
b = math.sqrt((x3-x1)**2+(y3-y1)**2)
a = math.sqrt((x3-x2)**2+(y3-y2)**2)
A = math.acos((a*a-b*b-c*c)/(-2*b*c))
B = math.acos((b*b-a*a-c*c)/(-2*a*c))
C = math.acos((c*c-b*b-a*a)/(-2*a*b))
print(math.degrees(A),math.degrees(B),math.degrees(C))
```

## 转义序列 \
- a = "He said,"Johon's program is easy to read"
- 转掉它原来的意思
- 一般情况下只有当语句与默认方法相撞的时候，就需要转义

```python
import hashlib
m='voodavdadvdvmv'
h= hashlib.md5()  #创建md5对象
h.update(m.encode(encoding='utf-8')) #生成加密串，其中password是要加密的字符串
print('加密前：'+m)  #打印经过md5加密的字符串
print('加密后: '+h.hexdigest())
```

```python
dataset = '54665a958bfa2158f239555ff628cf40'
password = input('>>')
h= hashlib.md5()
h.update(password.encode(encoding='utf-8'))
if dataset == h.hexdigest():
    print('success')
else:
    print('faild')
```
